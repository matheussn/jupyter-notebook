```python
%matplotlib inline
from matplotlib import pyplot as plt
from tensorflow.keras.utils import image_dataset_from_directory
import tensorflow as tf
```

```python
image_size = (64, 64)
batch_size = 9
path = '/home/neto/Documents/Linnaeus/train/'
```

```python
train_ds = image_dataset_from_directory(path, image_size=image_size, batch_size=batch_size)
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
```

```python
for j in range(10):
    normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
    imgs, labels = next(iter(normalized_ds))
    fig, axs = plt.subplots(3, 3)
    labels = labels.numpy()

    class_names = train_ds.class_names
    cnt = 0
    for i in range(3):
        for j in range(3):
#             img = (imgs[cnt] + 1) / 2.0
            axs[i, j].imshow(imgs[cnt])
            axs[i, j].axis('off')
            axs[i, j].set_title(f"Class: ${class_names[labels[cnt]]}")
            cnt += 1
```
